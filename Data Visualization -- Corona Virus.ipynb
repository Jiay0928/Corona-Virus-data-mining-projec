### Pie Graph and Bar


```python
Updatedf = df.loc[df['date'] == "2020-03-17" ]
CaPopu = 37644966
USpopu = 330441184
Chinapopu = 1437722468
Itapopu =60550075
labels = ['Canada','US','Italy','China']
CaseUpdate = Updatedf['total_cases']
CanadaPer = CaseUpdate.index[0]/CaPopu * 100

USPer = CaseUpdate.index[3]/USpopu * 100
ChinaPer = CaseUpdate.index[1]/Chinapopu * 100
ItaPer = CaseUpdate.index[2]/Itapopu * 100

nvalue = [CanadaPer,USPer,ItaPer,ChinaPer]

nlabels = ['Canada','US','Italy','China']
plt.title('Percentage in Different Country',fontdict = {'fontname': 'Comic Sans MS','fontsize' : 20})
plt.xlabel('Country')
plt.ylabel('Percentage (%)')
nbars = plt.bar(nlabels, nvalue)
```

### pie chart

```python
labels = ['Canada','China','Italy','United State']
values = Updatedf['new_cases']
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
patches, texts = plt.pie(values, colors=colors, explode = (0.1, 0, 0, 0),shadow=True, startangle=90)
plt.legend(patches, labels, loc="best")



plt.axis('equal')
plt.tight_layout()
plt.title ('New Cases',fontdict = {'fontname': 'Comic Sans MS','fontsize' : 20})
plt.show()
labels = ['Canada','China','Italy','United State']
values = Updatedf['total_cases']
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
patches, texts = plt.pie(values, explode = (0.1, 0, 0,0), colors=colors, shadow=True, startangle=90)
plt.legend(patches, labels, loc="best")
plt.axis('equal')
plt.tight_layout()
plt.title ('Total Case',fontdict = {'fontname': 'Comic Sans MS','fontsize' : 20})
plt.show()
```
